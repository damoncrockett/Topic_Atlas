```python
tmp = pd.DataFrame({"city":df.city.unique()})
```

```python
tmp = tmp[tmp.city.isin(top64)]
```

```python
tmp['vmeandiff'] = 0
tmp['vstddiff'] = 0
tmp['smeandiff'] = 0
tmp['sstddiff'] = 0
```

```python
for city in tmp.city:
    idx = tmp.index[tmp.city==city]
    
    vmean12 = df.meanval[(df.city==city)&(df.year==2012)].iloc[0] 
    vmean14 = df.meanval[(df.city==city)&(df.year==2014)].iloc[0]
    vstd12 = df.stdval[(df.city==city)&(df.year==2012)].iloc[0]
    vstd14 = df.stdval[(df.city==city)&(df.year==2014)].iloc[0]
    smean12 = df.meansat[(df.city==city)&(df.year==2012)].iloc[0]
    smean14 = df.meansat[(df.city==city)&(df.year==2014)].iloc[0]
    sstd12 = df.stdsat[(df.city==city)&(df.year==2012)].iloc[0]
    sstd14 = df.stdsat[(df.city==city)&(df.year==2014)].iloc[0]
    
    vmeandiff = abs( vmean14 - vmean12 )
    vstddiff = abs( vstd14 - vstd12 )
    smeandiff = abs( smean14 - smean12 ) 
    sstddiff = abs( sstd14 - sstd12 )
    
    tmp.vmeandiff.loc[idx] = vmeandiff
    tmp.vstddiff.loc[idx] = vstddiff
    tmp.smeandiff.loc[idx] = smeandiff
    tmp.sstddiff.loc[idx] = sstddiff
```

```python
hentropy = []
for i in range(len(df)):
    huevec = df.huevec.iloc[i]
    huevec = [float(item) for item in huevec.split("_")]
    hentropy.append(entropy(huevec))
```

```python
df['hentropy'] = hentropy
```

```python
df[(df.year==2014)&(df.city.isin(top64))].sort_values(by='hentropy').head()
```

```python
df[(df.year==2014)&(df.city.isin(top64))].sort_values(by='hentropy').tail()
```

```python
tmp['hentropydiff'] = 0
```

```python
for city in tmp.city:
    idx = tmp.index[tmp.city==city]
    
    hent12 = df.hentropy[(df.city==city)&(df.year==2012)].iloc[0] 
    hent14 = df.hentropy[(df.city==city)&(df.year==2014)].iloc[0]
    
    hentdiff = abs( hent14 - hent12 )

    
    tmp.hentropydiff.loc[idx] = hentdiff
```

```python
tmp.sort_values(by='hentropydiff').tail()
```

```python
tmp['kld'] = 0
```

```python
for city in tmp.city:
    idx = tmp.index[tmp.city==city]
    
    hv12 = df.huevec[(df.city==city)&(df.year==2012)].iloc[0] 
    hv14 = df.huevec[(df.city==city)&(df.year==2014)].iloc[0]
    
    hv12 = [float(item) for item in hv12.split("_")]
    hv14 = [float(item) for item in hv14.split("_")]
    
    kld = entropy(hv14,hv12)
    
    tmp.kld.loc[idx] = kld
```
