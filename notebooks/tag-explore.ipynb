# Exploring Machine Tags using Direct Visualization

#### Load and process data

```python
DIR = "/Users/damoncrockett/Desktop/sample/"
```

```python
md = pd.read_csv(DIR+"tagged-sample.csv")
```

```python
fd = pd.read_csv(DIR+"sample-vgg16.csv")
```

```python
md = md[md.hue_peak!='unreadable']
md.hue_peak = md.hue_peak.apply(int)
md.val_mean = md.val_mean.apply(float)
md.sat_mean = md.sat_mean.apply(float)
fd = fd.loc[md.index]
```

#### Animated Sampling

```python
for i in range(10):
    canvas = dv.montage(md.macpath,fd,'stage',thumb=64,sample=100)
    clear_output()
    display(canvas)
    time.sleep(2)
```
